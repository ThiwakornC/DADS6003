```python
colums_delete = ['Timestamp','จากข้อที่แล้ว เนื่องจากสาเหตุใด']
df_encoded = df_encoded.drop(columns=colums_delete, axis=1)
df_encoded
```

```python
columns = df_encoded.columns.tolist()
columns
```

```python
columns = ['เพศของผู้ตอบแบบสอบถาม',
 'ช่วงอายุของผู้ตอบแบบสอบถาม',
 'อาชีพของผู้ตอบแบบสอบถาม',
 'ช่วงเงินเดือนของผู้ตอบแบบสอบถาม',
 'ท่านใช้สมาร์ทโฟนในการทำอะไรมากที่สุด [เล่นเกม]',
 'ท่านใช้สมาร์ทโฟนในการทำอะไรมากที่สุด [เล่นโซเชี่ยลมีเดีย]',
 'ท่านใช้สมาร์ทโฟนในการทำอะไรมากที่สุด [ดูหนัง ฟังเพลง]',
 'ท่านใช้สมาร์ทโฟนในการทำอะไรมากที่สุด [ถ่ายรูป ถ่ายวิดีโอ]',
 'หากท่านซื้อสมาร์ทโฟนเครื่องใหม่ จะใช้เลือกซื้อสมาร์ทโฟนแบรนด์ใด']
df_encoded = pd.get_dummies(df_encoded, columns=columns)
df_encoded
```

```python
df_encoded = df_encoded[ df_encoded["ปัจจุบันท่านใช้สมาร์ทโฟนระบบปฎิบัติการใด"] != "IOS และ Android" ]
df_encoded
```

```python
df_encoded["ปัจจุบันท่านใช้สมาร์ทโฟนระบบปฎิบัติการใด"] = df_encoded["ปัจจุบันท่านใช้สมาร์ทโฟนระบบปฎิบัติการใด"].replace({"IOS": 1, 'Android': 0})
df_encoded
```

```python
df_encoded.info()
```

```python
df_encoded  = df_encoded.replace({True: 1, False: 0})
df_encoded
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Select features (X) and target (y)
X = df_encoded.drop('ปัจจุบันท่านใช้สมาร์ทโฟนระบบปฎิบัติการใด', axis=1)
y = df_encoded['ปัจจุบันท่านใช้สมาร์ทโฟนระบบปฎิบัติการใด']

y
```

```python

# Split the dataset into training and testing sets (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)

# Create a logistic regression model
logistic_model = LogisticRegression()

# Train the model
logistic_model.fit(X_train, y_train)

# Test the model
accuracy = logistic_model.score(X_test, y_test)

accuracy
```
